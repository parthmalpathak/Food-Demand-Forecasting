# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

# Load Data and Data Analysis

```python
#Load Data
df_train = pd.read_csv("train_GzS76OK/train.csv")
df_test = pd.read_csv("test_QoiMO9B.csv")
df_meal = pd.read_csv("train_GzS76OK/meal_info.csv")
df_fulfill = pd.read_csv("train_GzS76OK/fulfilment_center_info.csv")
df_sample = pd.read_csv("sample_submission_hSlSoT6.csv")
```

```python
# Display Data
display(df_train)
display(df_meal)
display(df_fulfill)
display(df_test)
display(df_sample)
```

```python
# [1] indicating the number of customers who have applied Promotional Discount
df_emailerPromotion = df_train.loc[df_train["emailer_for_promotion"].isin([1])]

print("Number of Customers visited with Promotion applied:", df_emailerPromotion["meal_id"].count())
```

```python
plt.plot(df_train["week"], df_train["num_orders"], c="b", label = "Number of orders every week")
plt.plot(df_emailerPromotion["week"], df_emailerPromotion["num_orders"], c="r", label = "Number of orders when promotion is applied")
plt.xlabel("Week")
plt.ylabel("Number of Orders")
plt.legend()
plt.title("Number of Orders per week")
plt.show()
```

```python
# [1] indicating the number of customers who purchased meal featured on homepage
df_homepageFeatured = df_train.loc[df_train["homepage_featured"].isin([1])]

print("Number of Customers who purchased meal featured on homepage:", df_homepageFeatured["meal_id"].count())
```

```python
plt.plot(df_train["week"], df_train["num_orders"], c="b", label = "Number of orders every week")
plt.plot(df_homepageFeatured["week"], df_homepageFeatured["num_orders"], c="r", label = "Number of orders featured on homepage")
plt.xlabel("Week")
plt.ylabel("Number of Orders")
plt.legend()
plt.title("Number of Orders per week")
plt.show()
```

```python
#Check Base and Sale Price Difference
priceDiff = df_train["base_price"] - df_train["checkout_price"]
df_train["price_difference"] = priceDiff
df_train
```

```python
#Meal Data Analysis
lst_cuisine = list(df_meal["cuisine"].value_counts())
lst_category = list(df_meal["category"].value_counts())

plt.figure(1)
plt.pie(lst_cuisine, labels = df_meal["cuisine"].unique())
plt.figure(2)
plt.pie(lst_category, labels = df_meal["category"].unique())
plt.show()
```

```python
df_train = df_train.drop(["price_difference"], axis = 1)
df_train
```

```python
#Features and Target Split
X = df_train.drop(["num_orders"], axis = 1)
Y = df_train["num_orders"]
```

```python
#Train test Split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
from sklearn.feature_selection import mutual_info_regression, mutual_info_classif
regress_importances = mutual_info_regression(X_train, y_train)
```

```python
headerInfo = list(X_train.columns.values)
plt.figure()
f, ax = plt.subplots(figsize=(15,7))
plt.bar(headerInfo, regress_importances)
plt.xlabel("Data Headers")
plt.ylabel("Mutual Regression Information")
plt.title("Mutual Regression Infomation with num_orders")
plt.show()
```

# Machine Learning Prediction: Base Model

```python
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, AdaBoostRegressor
import xgboost
from sklearn.metrics import mean_squared_log_error
```

```python
params = {
    "n_estimators": 500,
    "max_depth": 4,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "squared_error",
}

gb = GradientBoostingRegressor(**params)
gb.fit(X_train, y_train)
gb_preds = gb.predict(X_test)
gb_preds
```

```python
print("Root Mean Squared Log Error:", np.sqrt(mean_squared_log_error(y_test, gb_preds)))
```

```python
xgb = xgboost.XGBRegressor(learning_rate =.1, n_estimators=10,
                     max_depth=2, min_child_weight=3, gamma=0, 
                     subsample=.8, colsample_bytree=.7, reg_alpha=1, 
                     objective= 'reg:squarederror')
xgb.fit(X_train, y_train)
xgb_preds = xgb.predict(X_test)
xgb_preds
```

```python
print("Root Mean Squared Log Error:", np.sqrt(mean_squared_log_error(y_test, xgb_preds)))
```

```python
regr = AdaBoostRegressor(random_state=0, n_estimators=500)
regr.fit(X_train,y_train)
regr_preds = regr.predict(X_test)
regr_preds
```

```python
print("Root Mean Squared Log Error:", np.sqrt(mean_squared_log_error(y_test, regr_preds)))
```

# Feature Engineering

```python
df_train_center_id = pd.merge(df_train, df_fulfill, on = 'center_id')
df_test_center_id = pd.merge(df_test, df_fulfill, on = 'center_id')
display(df_train_center_id.head())
display(df_test_center_id.head())
```

```python
df_train_final = pd.merge(df_train_center_id, df_meal, on = 'meal_id')
df_test_final = pd.merge(df_test_center_id, df_meal, on = 'meal_id')
display(df_train_final.head())
display(df_test_final.head())
```

```python
df_train_final.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_train_final['center_type_encoded'] = label_encoder.fit_transform(df_train_final['center_type'])
df_train_final['cuisine_encoded'] = label_encoder.fit_transform(df_train_final['cuisine'])
df_train_final['category_encoded'] = label_encoder.fit_transform(df_train_final['category'])

df_test_final['center_type_encoded'] = label_encoder.fit_transform(df_test_final['center_type'])
df_test_final['cuisine_encoded'] = label_encoder.fit_transform(df_test_final['cuisine'])
df_test_final['category_encoded'] = label_encoder.fit_transform(df_test_final['category'])
```

```python
df_train_featurized = df_train_final.drop(["center_type", "cuisine", "category"], axis = 1)
display(df_train_featurized.head())
print(df_train_featurized.dtypes)

df_test_featurized = df_test_final.drop(["center_type", "cuisine", "category"], axis = 1)
display(df_test_featurized.head())
print(df_test_featurized.dtypes)
```

```python
column_correlation = df_train_featurized.corr()
display(column_correlation)
sns.heatmap(column_correlation)
```

```python
X_updated = df_train_featurized.drop(["num_orders"], axis = 1)
Y_updated = df_train_featurized["num_orders"]
X_updated_train, X_updated_test, y_updated_train, y_updated_test = train_test_split(X_updated, Y_updated, test_size=0.25, random_state=0)
display(X_updated_train.values.shape)
display(y_updated_train.values.shape)
display(X_updated_test.values.shape)
```

# Machine Learning Prediction: Updated Data

```python
mm=RandomForestRegressor(n_jobs=-1,n_estimators=200,oob_score=True)
mm.fit(X_updated_train, y_updated_train)
mm_preds = mm.predict(X_updated_test)
np.sum(mm_preds<0)
```

```python
print("Root Mean Squared Log Error:", np.sqrt(mean_squared_log_error(y_updated_test, mm_preds)))
```

```python
solution = mm.predict(df_test_featurized)
```

# Convert Result to CSV

```python
soln_df = df_sample
soln_df["num_orders"] = solution
display(soln_df)
```

```python
Convert Solution File to CSV
soln_df.to_csv('food_forecasting_solution.csv', index=False)
```
